```python
# Choix du modèle de ML : Régression logistique, classification car variables discrètes
# choix des labels/features
y_col = ['LUNG_CANCER']
x_col = [c for c in df.columns if c not in y_col]

# Preprocessing:

# Pas de valeur nulle dans le dataset : pas besoin d'utiliser dropna

# Pas de valeur manquante dans le dataset : pas de valeur à remplacer

# valeurs discrètes partout dans le dataset "1" et "2" sauf dans "GENDER" et "LUNG CANCER" variables catégorielles "YES", "NO" donc/
# on va utiliser la copie df dataset df2

# Train test split : diviser en 2 le dataset pour entraîner le modèle
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[x_col], df[y_col], test_size=0.2, random_state=0)
# on garde 20% du dataset pour le test set

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

# Séparer les variables numériques et catégoriques
x_num_col = x_train.select_dtypes(include='number').columns.values
x_str_col = x_train.select_dtypes(include='object').columns.values

# remplacement des valeurs manquantes : ici pas nécessaire
# s1 = ColumnTransformer(transformers = [
    #('imputer_num', SimpleImputer(strategy='mean'), x_num_col),
    #('imputer_str', SimpleImputer(strategy='most_frequent'), x_str_col)], verbose_feature_names_out=False)

# Mise à l'échelle et encodage
s2 = ColumnTransformer(transformers = [
     ('scaler', StandardScaler(), x_num_col),
    ('encoder', OneHotEncoder(drop='first'), x_str_col)], verbose_feature_names_out=False)

# application au train set
# x_train = pd.DataFrame(s1.fit_transform(x_train), columns=s1.get_feature_names_out())
x_train = pd.DataFrame(s2.fit_transform(x_train), columns=s2.get_feature_names_out())
# application au test set
# x_test = pd.DataFrame(s1.transform(x_test), columns=s1.get_feature_names_out())
x_test = pd.DataFrame(s2.transform(x_test), columns=s2.get_feature_names_out())

# Encodage des labels (rappel : pas de valeur manquante)
y_num_col = y_train.select_dtypes(include='number').columns.values
y_str_col = y_train.select_dtypes(include='object').columns.values

# On utilise OrdinalEncoder
s3 = ColumnTransformer(transformers = [
    ('scaler', StandardScaler(), y_num_col),
    ('encoder', OrdinalEncoder(), y_str_col)], verbose_feature_names_out=False, remainder='passthrough')

y_train = pd.DataFrame(s3.fit_transform(y_train), columns=s3.get_feature_names_out())
y_test = pd.DataFrame(s3.transform(y_test), columns=s3.get_feature_names_out())

x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
# entraînement du modele de régression logistique

classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train.iloc[:,0])
y_pred = classifier.predict(x_train)

# Par défaut, sklearn utilise l'accuracy comme score
print(classifier.score(x_train, y_train.iloc[:,0]))
# On examine les résultats individuellement
tp = len(y_pred[(y_pred == 1.) & (y_train.iloc[:,0] == 1.)]) #true positive
tn = len(y_pred[(y_pred == 0.) & (y_train.iloc[:,0] == 0.)]) #true negative
fp = len(y_pred[(y_pred == 1.) & (y_train.iloc[:,0] == 0.)]) #false positive
fn = len(y_pred[(y_pred == 0.) & (y_train.iloc[:,0] == 1.)]) #false negative
print(tp,tn,fp,fn)
```

```python
# l'accuracy est excellente à 0.939 = le modèle a prédit correctement "LUNG_CANCER"
# le modèle prédit correctement 213 vrais positifs (tp)
# il prédit de même correctement 19  vrais négatifs (tn)
# il prédit 10 faux positifs et 5 faux négatifs
```

```python
tn, fp, fn, tp = confusion_matrix(y_train.iloc[:,0], y_pred).ravel() # fonctionne seulement avec 2 catégories, sinon plus de 4 chiffres à sortir
```

```python
# matrice de confusion du train set
sns.heatmap(confusion_matrix(y_train.iloc[:,0], y_pred), annot=True, fmt='d').set(xlabel="Predicted value", ylabel="True value", title="Train");
```

```python
# calcul des autres métriques sur le train set
print("accuracy\t%f" % ((tp+tn)/(tp+tn+fp+fn)))
print("sensitivity\t%f" % (tp/(tp+fn))) # mesure avec quelle sensibilité le modèle détecte les cas positifs (= éviter de faire des faux négatifs !)
print("specificity\t%f" % (tn/(tn+fp))) # mesure le degré de spécificité : détecte les cas négatifs (évite les faux positifs)
print("precision\t%f" % (tp/(tp+fp)))
```

```python
# la sensibilité au vrai positif est excellente
# la spécificité (détection des vrais négatifs) un peu moins bonne
```

```python
# Maintenant même calcul sur le test set
y_pred = classifier.predict(x_test)
tn, fp, fn, tp = confusion_matrix(y_test.iloc[:,0], y_pred).ravel()
print("accuracy   \t%f" % ((tp+tn)/(tp+tn+fp+fn)))
print("sensitivity\t%f" % (tp/(tp+fn)))
print("specificity\t%f" % (tn/(tn+fp)))
print("precision  \t%f" % (tp/(tp+fp)))
# matrice de confusion du test set
sns.heatmap(confusion_matrix(y_test.iloc[:,0], y_pred), annot=True, fmt='d').set(xlabel="Predicted value", ylabel="True value", title="Test");
```

```python
# la spécificité descend à 0.5 = le modèle n'a correctement identifié que la moitié des cas de "non cancer"
# la précision est de 0.91 cad que parmi tous les cas identifiés comme positifs par le modèle, 91% le sont vraiment.
```

```python
from scipy.stats import chi2_contingency

# Analyse des contingences entre les variables du dataset et le diagnostic "LUNG_CANCER" à l'aide du test chi carré
# En effet les variables n'étant pas continues je ne peux pas utiliser de ztest ou de heatmap

# Liste des variables catégoriques du dataset
categorical_vars = categorical_vars = df.select_dtypes(include=['object', 'category',"int64"]).columns.tolist()

# Création d'une liste pour stocker les résultats des tests du chi carré
chi2_results = []

# Une boucle for sur chaque variable catégorielle :
for var in categorical_vars:
  if var != "LUNG_CANCER":
# Création du tableau de contingence
    contingency_table = pd.crosstab(df['LUNG_CANCER'], df[var])

    # Affichage du tableau de contingence
    print(f"Tableau de contingence pour {var}:\n{contingency_table}\n")

    # Test du chi carré
    chi2, p_value, _, _ = chi2_contingency(contingency_table)

    # Stockage des résultats
    chi2_results.append({'Variable': var, 'Chi2': chi2, 'P-value': p_value})

    # Affichage des résultats du test du chi carré
    print(f"Résultats du test du chi carré pour {var}:")
    print(f"Chi2: {chi2}")
    print(f"P-value: {p_value}\n")

# Conversion des résultats en Dataframe
chi2_results_df = pd.DataFrame(chi2_results)

# Affichage graphique des résultats du test du chi carré
plt.figure(figsize=(10, 6))
plt.bar(chi2_results_df['Variable'], chi2_results_df['P-value'])
plt.xlabel('Variable')
plt.ylabel('P-value')
plt.title('Résultats du test du chi carré pour chaque variable')
plt.xticks(rotation=45)
plt.show()
```
